```python
# a. Aplica la transformación SVD para obtener la matriz de las variables
#    latentes de los restaurantes, en relación a la manera en que los usuarios
#    evaluaron la comida. Deberás obtener la factorización SVD con el máximo número
#    de componentes y por lo tanto el máximo de valores singulares posibles de acuerdo
#    a este criterio. Así, de acuerdo al resultado obtenido en el Ejercicio-3, asigna
#    a la variable "nc_food" el número entero adecuado para llegar a obtener el
#    resultado pedido.


# ************* Inlcuye aquí tu código:**********************************************

import pandas as pd

data1 = pd.read_csv("rating_final.csv", header='infer', sep=",")
data2 = pd.read_csv("geoplaces2.csv", header='infer', encoding='latin-1')

lista_data1 = ['userID', 'placeID', 'food_rating']
lista_data2 = ['placeID', 'name']

# DataFrames filtrados con las columnas necesarias
data1a = data1[lista_data1]
data2a = data2[lista_data2]

# Combinar los DataFrames a través de la columna común "placeID"
df_combinado = pd.merge(data1a, data2a, on='placeID')

# Crear la matriz de utilidad UtMx_food
UtMx_food = df_combinado.pivot_table(index='name', columns='userID', values='food_rating')

# Mostrar las primeras filas de la matriz de utilidad para verificar
print("Primeras filas de UtMx_food:")
print(UtMx_food.head())

import numpy as np
from scipy.linalg import svd

# Reemplazar NaN con 0 en la matriz de utilidad
UtMx_food_filled = UtMx_food.fillna(0)

# Convertir a matriz NumPy
A = UtMx_food_filled.values

# Aplicar SVD
U, Sigma, VT = svd(A)

# Determinar el número de componentes (valores singulares)
nc_food = min(A.shape)

# Mostrar los resultados de SVD
print("Matriz U:")
print(U)
print("Valores singulares (Sigma):")
print(Sigma)
print("Matriz VT:")
print(VT)

nc_food = min(UtMx_food.shape)
print("Número de componentes (nc_food):", nc_food)




# *********** Aquí termina la sección de agregar código *****************************

print('Total de valores singulares basados en evaluación de la comida:', nc_food)
```

```python
# b. Lleva a cabo la factorización SVD truncada en la matriz de utilidad
#    obtenida previamente y con respecto al número de componentes dada
#    por la variable nc_food.


# ************* Inlcuye aquí tu código:********************************

import pandas as pd

data1 = pd.read_csv("rating_final.csv", header='infer', sep=",")
data2 = pd.read_csv("geoplaces2.csv", header='infer', encoding='latin-1')

# Definir las listas con las columnas necesarias
lista_data1 = ['userID', 'placeID', 'food_rating']
lista_data2 = ['placeID', 'name']

# Crear los DataFrames filtrados con las columnas necesarias
data1a = data1[lista_data1]
data2a = data2[lista_data2]

# Combinar los DataFrames a través de la columna común "placeID"
df_combinado = pd.merge(data1a, data2a, on='placeID')

# Crear la matriz de utilidad UtMx_food
UtMx_food = df_combinado.pivot_table(index='name', columns='userID', values='food_rating')

# Reemplazar NaN con 0 en la matriz de utilidad
UtMx_food_filled = UtMx_food.fillna(0)

# Mostrar las primeras filas de la matriz de utilidad para verificar
print("Primeras filas de UtMx_food:")
print(UtMx_food.head())

# Calcular el número de componentes
nc_food = min(UtMx_food.shape) - 1
print("Número de componentes (nc_food):", nc_food)

import numpy as np
from scipy.sparse.linalg import svds

# Convertir a matriz NumPy
A = UtMx_food_filled.values

# Aplicar SVD truncada
U, Sigma, VT = svds(A, k=nc_food)

# Crear la matriz SVD_food con las matrices U, Sigma y VT truncadas
SVD_food = (U, Sigma, VT)

# Mostrar los resultados de SVD truncada
print("Matriz U truncada:")
print(U)
print("Valores singulares truncados (Sigma):")
print(Sigma)
print("Matriz VT truncada:")
print(VT)


# *********** Aquí termina la sección de agregar código ***************


# Con la factorización SVD truncada obtenida, determinemos la cantidad
# de componentes que explican un 90% de la variabilidad acumulada de
# cada componente:

explained_variance = np.square(Sigma) / np.sum(np.square(Sigma))

explained_variance_ratio_ = np.cumsum(explained_variance)

# Determinar la cantidad de componentes que explican el 90% de la variabilidad acumulada
for j in range(nc_food):
    if explained_variance_ratio_[0:j].sum() > 0.90:
        break

# Usar esta cantidad de componentes para las recomendaciones basadas en la calificación de la comida
N_food = j-1

print('Valor de truncamiento al 90% de dicha variabilidad:', N_food)


```
